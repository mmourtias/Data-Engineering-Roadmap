## 1. Î•Î¾ÎµÏÎµÏÎ½Î·ÏƒÎ· Ï„Î¿Ï… DataFrame
Î ÏÎ¹Î½ Î¾ÎµÎºÎ¹Î½Î®ÏƒÎ¿Ï…Î¼Îµ Ï„Î·Î½ ÎµÏ€ÎµÎ¾ÎµÏÎ³Î±ÏƒÎ¯Î±, Ï€ÏÎ­Ï€ÎµÎ¹ Î½Î± ÎºÎ±Ï„Î±Î»Î¬Î²Î¿Ï…Î¼Îµ Ï„Î· Î´Î¿Î¼Î® Ï„Ï‰Î½ Î´ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½:
* `.head()`: Î•Ï€Î¹ÏƒÏ„ÏÎ­Ï†ÎµÎ¹ Ï„Î¹Ï‚ Ï€ÏÏÏ„ÎµÏ‚ Î»Î¯Î³ÎµÏ‚ Î³ÏÎ±Î¼Î¼Î­Ï‚.
* `.info()`: Î”ÎµÎ¯Ï‡Î½ÎµÎ¹ Î¿Î½ÏŒÎ¼Î±Ï„Î± ÏƒÏ„Î·Î»ÏÎ½, Ï„ÏÏ€Î¿Ï…Ï‚ Î´ÎµÎ´Î¿Î¼Î­Î½Ï‰Î½ (Dtype) ÎºÎ±Î¹ Î±Î½ Ï…Ï€Î¬ÏÏ‡Î¿Ï…Î½ ÎºÎµÎ½Î­Ï‚ Ï„Î¹Î¼Î­Ï‚ (non-null).
* `.shape`: Î•Ï€Î¹ÏƒÏ„ÏÎ­Ï†ÎµÎ¹ Ï„Î¿Î½ Î±ÏÎ¹Î¸Î¼ÏŒ Î³ÏÎ±Î¼Î¼ÏÎ½ ÎºÎ±Î¹ ÏƒÏ„Î·Î»ÏÎ½ (attribute, Ï‡Ï‰ÏÎ¯Ï‚ Ï€Î±ÏÎµÎ½Î¸Î­ÏƒÎµÎ¹Ï‚).
* `.describe()`: Î¥Ï€Î¿Î»Î¿Î³Î¯Î¶ÎµÎ¹ Î²Î±ÏƒÎ¹ÎºÎ¬ ÏƒÏ„Î±Ï„Î¹ÏƒÏ„Î¹ÎºÎ¬ (mean, std, min, max) Î³Î¹Î± Î±ÏÎ¹Î¸Î¼Î·Ï„Î¹ÎºÎ­Ï‚ ÏƒÏ„Î®Î»ÎµÏ‚.
* `.values`: Î¤Î± Î´ÎµÎ´Î¿Î¼Î­Î½Î± ÏƒÎµ Î¼Î¿ÏÏ†Î® NumPy array.
* `.columns`: Î¤Î± Î¿Î½ÏŒÎ¼Î±Ï„Î± Ï„Ï‰Î½ ÏƒÏ„Î·Î»ÏÎ½.
* `.index`: Î¤Î± labels Ï„Ï‰Î½ Î³ÏÎ±Î¼Î¼ÏÎ½ (row numbers).


```python
import pandas as pd

# Î”Î·Î¼Î¹Î¿Ï…ÏÎ³Î¯Î± ÎµÎ½ÏŒÏ‚ test DataFrame (ÏƒÎ±Î½ Î±Ï…Ï„ÏŒ Î¼Îµ Ï„Î± ÏƒÎºÏ…Î»Î¹Î¬)
data = {
    "name": ["Bella", "Charlie", "Lucy", "Cooper"],
    "breed": ["Labrador", "Poodle", "Chow Chow", "Schnauzer"],
    "color": ["Black", "Brown", "White", "Gray"],
    "height_cm": [56, 43, 46, 49],
    "weight_kg": [25, 10, 22, 17],
    "date_of_birth": ["2013-07-01", "2016-09-16", "2014-08-25", "2011-12-11"]
}
dogs = pd.DataFrame(data)

# Î”Î¿ÎºÎ¯Î¼Î±ÏƒÎµ Ï„Î¹Ï‚ ÎµÎ½Ï„Î¿Î»Î­Ï‚
print(dogs.head())
print(dogs.info())
print(dogs.shape)
```

## 2. Î¤Î±Î¾Î¹Î½ÏŒÎ¼Î·ÏƒÎ· ÎºÎ±Î¹ Subsetting
### Î¤Î±Î¾Î¹Î½ÏŒÎ¼Î·ÏƒÎ· (Sorting)
* `df.sort_values("column")`: Î¤Î±Î¾Î¹Î½ÏŒÎ¼Î·ÏƒÎ· Î±ÏÎ¾Î¿Ï…ÏƒÎ±.
* `ascending=False`: Î“Î¹Î± Ï†Î¸Î¯Î½Î¿Ï…ÏƒÎ± Ï„Î±Î¾Î¹Î½ÏŒÎ¼Î·ÏƒÎ·.

### Î•Ï€Î¹Î»Î¿Î³Î® Î£Ï„Î·Î»ÏÎ½ & Î“ÏÎ±Î¼Î¼ÏÎ½ (Subsetting)
* `df["column"]`: Î•Ï€Î¹Î»Î¿Î³Î® Î¼Î¯Î±Ï‚ ÏƒÏ„Î®Î»Î·Ï‚.
* `df[["col1", "col2"]]`: Î•Ï€Î¹Î»Î¿Î³Î® Ï€Î¿Î»Î»Î±Ï€Î»ÏÎ½ ÏƒÏ„Î·Î»ÏÎ½ (Î´Î¹Ï€Î»Î­Ï‚ Î±Î³ÎºÏÎ»ÎµÏ‚).
* `df[df["col"] > x]`: Î¦Î¹Î»Ï„ÏÎ¬ÏÎ¹ÏƒÎ¼Î± Î³ÏÎ±Î¼Î¼ÏÎ½ Î¼Îµ Î²Î¬ÏƒÎ· ÏƒÏ…Î½Î¸Î®ÎºÎ·.
* `.isin([list])`: Î¦Î¹Î»Ï„ÏÎ¬ÏÎ¹ÏƒÎ¼Î± Î³Î¹Î± Ï€Î¿Î»Î»Î±Ï€Î»Î­Ï‚ Ï„Î¹Î¼Î­Ï‚ ÏƒÎµ Î¼Î¯Î± ÏƒÏ„Î®Î»Î·.

```python
# 1. Î¤Î±Î¾Î¹Î½ÏŒÎ¼Î·ÏƒÎ· ÎºÎ±Ï„Î¬ Î²Î¬ÏÎ¿Ï‚ (Ï†Î¸Î¯Î½Î¿Ï…ÏƒÎ±)
dogs_heavy = dogs.sort_values("weight_kg", ascending=False)

# 2. Î¦Î¹Î»Ï„ÏÎ¬ÏÎ¹ÏƒÎ¼Î±: Î£ÎºÏ…Î»Î¹Î¬ Ï€Î¹Î¿ ÏˆÎ·Î»Î¬ Î±Ï€ÏŒ 45cm
tall_dogs = dogs[dogs["height_cm"] > 45]

# 3. Î§ÏÎ®ÏƒÎ· .isin(): ÎœÏŒÎ½Î¿ Labradors ÎºÎ±Î¹ Poodles
lab_poodle = dogs[dogs["breed"].isin(["Labrador", "Poodle"])]

print(lab_poodle)
```

## 3. Î ÏÎ¿ÏƒÎ¸Î®ÎºÎ· ÎÎ­Ï‰Î½ Î£Ï„Î·Î»ÏÎ½ (Mutating)
ÎœÏ€Î¿ÏÎ¿ÏÎ¼Îµ Î½Î± Î´Î·Î¼Î¹Î¿Ï…ÏÎ³Î®ÏƒÎ¿Ï…Î¼Îµ Î½Î­ÎµÏ‚ ÏƒÏ„Î®Î»ÎµÏ‚ ÎºÎ¬Î½Î¿Î½Ï„Î±Ï‚ Ï€ÏÎ¬Î¾ÎµÎ¹Ï‚ ÏƒÏ„Î¹Ï‚ Ï…Ï€Î¬ÏÏ‡Î¿Ï…ÏƒÎµÏ‚.
* `df["new_col"] = df["old_col"] / 100`

```python
# Î ÏÎ¿ÏƒÎ¸Î®ÎºÎ· ÏÏˆÎ¿Ï…Ï‚ ÏƒÎµ Î¼Î­Ï„ÏÎ±
dogs["height_m"] = dogs["height_cm"] / 100

# Î¥Ï€Î¿Î»Î¿Î³Î¹ÏƒÎ¼ÏŒÏ‚ BMI (weight / height^2)
dogs["bmi"] = dogs["weight_kg"] / (dogs["height_m"] ** 2)

print(dogs.head())
```
